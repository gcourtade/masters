```python
import nmrglue as ng
import numpy as np
import matplotlib.pyplot as plt
```

```python
#This code cell makes a plot of B22 spectrum with zoomed in regions

#data input:
nmr_file='1/pdata/1'
dic, data = ng.bruker.read_pdata(nmr_file)

#convert to pipe
C = ng.convert.converter()
C.from_bruker(dic, data)
pdic,pdata = C.to_pipe()

#ppm scale
uc = ng.pipe.make_uc(pdic,pdata)

#plot:
fig = plt.figure(figsize=(12,12))
gs = fig.add_gridspec(2, 1, height_ratios=[2.5, 1])
axs = gs.subplots()

for i in range(0,2):
    axs[i].plot(uc.ppm_scale(), data, 'k-', linewidth=0.8)
    axs[i].invert_xaxis() #invert chemical shift scale
    axs[i].set_xlabel("Chemical shift (ppm)")
    axs[i].set_ylabel("Signal (A.U.)")
    axs[i].spines['top'].set_visible(False)
    axs[i].spines['right'].set_visible(False)



#whole spectrum
axs[0].set_xlim(11.2, -1)
axs[0].set_ylim(-2e4, 2e5)

#Trp region
axs[1].set_xlim(11.2, 9.5)
axs[1].set_ylim(-1e3, 9e3)

plt.savefig('B22_NMR.png', dpi=300, transparent=True)
plt.show()

plt.close()
```

```python
#This code cell makes plot of B22 with BHET with zoomed in regions 

#data input:
nmr_file='3/pdata/1'
dic, data = ng.bruker.read_pdata(nmr_file)

#convert to pipe
C = ng.convert.converter()
C.from_bruker(dic, data)
pdic,pdata = C.to_pipe()

#ppm scale
uc = ng.pipe.make_uc(pdic,pdata)

#Plot:
fig = plt.figure(figsize=(12,12))
gs = fig.add_gridspec(2, 1, height_ratios=[2.5, 1])
axs = gs.subplots()

for i in range(0,2):
    axs[i].plot(uc.ppm_scale(), data, color='steelblue', linewidth=0.8)
    axs[i].invert_xaxis() #invert chemical shift scale
    axs[i].set_xlabel("Chemical shift (ppm)")
    axs[i].set_ylabel("Signal (A.U.)")
    axs[i].spines['top'].set_visible(False)
    axs[i].spines['right'].set_visible(False)


#whole spectrum
axs[0].set_xlim(11.2, -1)
axs[0].set_ylim(-2e4, 2e5)

#Trp region
axs[1].set_xlim(11.2, 9.5)
axs[1].set_ylim(-1e3, 9e3)

plt.savefig('B22_NMR_BHET.png', dpi=300, transparent=True)
plt.show()

plt.close()
```

```python
#This code cell makes a plot of peaks corresponding to BHET, MHET, TPA and TSP for the NMR-assignment

nmr_file='64/pdata/107'
dic, data = ng.bruker.read_pdata(nmr_file)

#convert to pipe
C = ng.convert.converter()
C.from_bruker(dic, data)
pdic,pdata = C.to_pipe()

#ppm scale
uc = ng.pipe.make_uc(pdic,pdata)

#plot:
fig, axs = plt.subplots(1,2, constrained_layout=True, figsize=(12, 6))

for i in range(0,2):
    axs[i].plot(uc.ppm_scale(), data, 'k-', linewidth=0.8)
    axs[i].invert_xaxis()
    axs[i].spines['top'].set_visible(False)
    axs[i].spines['right'].set_visible(False)
    axs[i].spines['left'].set_visible(False)
    axs[i].set(ylabel=None)
    axs[i].set(yticklabels=[])  
    axs[i].tick_params(left=False)
    

#Show BHET, MHET, TPA
axs[0].set_xlim(8.3, 7.9)
axs[0].set_ylim(-3e3, 4.25e5)

#Show TSP
axs[1].set_xlim(0.3, -0.1)
axs[1].set_ylim(-3e3, 4.25e5)


fig.text(0.5, -0.03, "Chemical shift (ppm)", fontsize='10', ha='center')
plt.savefig('NMRspectra_timeresolved.png', dpi=300)         
plt.show()
```
